```python
import spacy
```

```python
model = spacy.load('trained/model-last/')
```

```python
test_data = ''
with open('test/test_req.txt', 'r') as rfile:
    test_data = rfile.read()
test_data = test_data.split('---')
```

```python
from cleaner import clean_result
import pandas as pd

df = pd.DataFrame(columns=['Text', 'Entities'])
```

```python
for task in test_data:
    doc = model(task)
    target_list = list(map(lambda _:_.text, doc.ents))
    print(doc.ents[0].start, doc.ents[0].end, doc.ents[0].label_, doc.ents[0].start_char)
    print(target_list)
```

```python
from spacy import displacy

parser = spacy.load("ru_core_news_lg")

doc_parsed = parser(test_data[0])
options = {"bg": "white", "distance": 130,
           "color": "black", "font": "Source Sans Pro"}
displacy.render(doc_parsed, style="dep", options=options)
```

```python
import dp.parser as parser
```

```python
doc = model(test_data[0])
for entity in parser.parse_text_for_ner(doc, doc.text):
    print(entity.name, [(attr.name, attr.type) for attr in entity.attributes])
```

```python
for task in test_data:
    doc = model(task)

    target_list = list(map(lambda _:_.text, doc.ents))
    result = clean_result.clean(target_list)
    df.loc[len(df.index)] = [task, result]
    print(task, '\n\n >>>> ', result, '\n\n')
```
